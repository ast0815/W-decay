```python
E_e = 0.45 * u.MeV
I = 1e-3 * u.A
#Q_EC = -382 * u.keV # 58Ni
#Q_EC = -190 * u.keV # 106Cd
Q_EC = -167 * u.keV # 35Cl
#Q_EC = -34 * u.keV # 107Ag
d = 1 * u.mm
n = 6e23 / u.cm3
```

```python
P = d * 2 * u.MeV / u.cm / c.eplus * I
P / u.kW
```

```python
# https://journals.aps.org/prd/pdf/10.1103/PhysRevD.60.053003
xsec = 0.0952 * ((E_e + Q_EC) / u.MeV)**2 * 1e-42 * u.cm2
xsec / u.cm2
```

```python
rate = xsec * (I / c.eplus) * d * n
```
